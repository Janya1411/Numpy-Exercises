## Seaborn - Visualizing Linear Relationships

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set(color_codes=True)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,x_jitter=.05)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,x_estimator=np.mean)
```

```python
anscombe = sns.load_dataset('anscombe')
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'I'"),ci=None,scatter_kws={'s': 80})
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'II'"),ci=None,scatter_kws={'s': 80})
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'II'"),ci=None,scatter_kws={'s': 80},order=2)
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'III'"),ci=None,scatter_kws={'s':80})
```

```python
sns.lmplot(x='x',y='y',data=anscombe.query("dataset == 'III'"),ci=None,scatter_kws={'s':80},robust=True)
```

```python
tips['big_tip'] = (tips.tip/tips.total_bill) > 0.15
```

```python
sns.lmplot(x= 'total_bill',y='big_tip',data=tips,y_jitter=.03)
```

```python
sns.lmplot(x= 'total_bill',y='big_tip',data=tips,y_jitter=.03,logistic=True)
```

```python
sns.residplot(x='x',y='y',data=anscombe.query("dataset =='I'"),scatter_kws={'s':80})
```

```python
sns.residplot(x='x',y='y',data=anscombe.query("dataset =='II'"),scatter_kws={'s':80})
```

```python
sns.lmplot(x='total_bill',y = 'tip', hue='smoker',data=tips)
```

```python
sns.lmplot(x='total_bill',y = 'tip', hue='smoker',data=tips,markers=['x','o'])
```

```python
sns.lmplot(x= 'total_bill',y='tip',hue='smoker',data=tips,col='time')
```

```python
sns.lmplot(x='total_bill',y='tip',hue='smoker',col='time',row='sex',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.pairplot(tips,x_vars=['total_bill','size'],y_vars=['tip'],height=5,aspect=0.8,kind='reg')
```
