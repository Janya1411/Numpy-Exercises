### Extracting data from SQL/Pulling Data from SQL Query

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite://',echo=False)
```

```python
df = pd.read_csv('example')
```

```python
df
```

```python
df.to_sql('data',engine)
```

```python
new_df = pd.read_sql('data',con=engine)
```

```python
new_df
```

```python
pd.read_sql_query('SELECT * FROM data', con=engine)
```

```python
pd.read_sql_query('SELECT b FROM data', con=engine)
```

```python
pd.read_sql_table('data',con=engine)
```
