### Missing and Cleaning Data

```python
import numpy as np
import pandas as pd
```

```python
url = 'http://rcs.bu.edu/examples/python/data_analysis/flights.csv'
```

```python
df = pd.read_csv(url,sep=',')
```

```python
df.sample(20)
```

```python
df_not_missing = df.dropna()
```

```python
df_not_missing.sample(20)
```

```python
df.dropna(axis=1,how='all')
```

```python
df.fillna(0)
```

```python
df.sample(20)
```

```python
df['dep_time'].fillna(df['dep_time'].mean())
```

```python
df['dep_time'].fillna(df.groupby('carrier')['dep_time'].transform('mean'))
```

```python
df[df['dep_time'].not_null() & df['dep_delay'].not_null()]
```

```python
df.isnull()
```

#### Case Study

```python
missing_df = df.isnull().sum(axis=0).reset_index()
missing_df.columns = ['column_name','missing_count']
```

```python
missing_df
```

```python
missing_df = missing_df.loc[missing_df['missing_count'] > 0]
missing_df = missing_df.sort_values(by='missing_count',ascending=False)
```
